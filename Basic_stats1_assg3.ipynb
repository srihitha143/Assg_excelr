# Descriptive Analytics for Numerical Columns:
It focuses on summarizing and describing the main features of a dataset.

```python
import pandas as pd
import numpy as np

#Load the dataset into a data analysis tool or programming environment.
file_path="/content/drive/MyDrive/Python excelr/sales_data_with_discounts.csv.xls"
df=pd.read_csv(file_path) #as ours is csv file , we are using read_csv()
print(df.head()) #.head() gives the quick snapshot of our data(first 5 rows)

#Identify numerical columns in the dataset.
numeric_df=df.select_dtypes(include=['int64','float64']) #here include=['int64','float64'] gives only columns with integer or floating point numbers
print("\n\n Numberic Column in the dataset:")
print("/n/n",numeric_df.columns)

#(a)Measures of Central Tendency and (b)Measures of Dispersion
#(a)Calculate the mean, median, mode, and (b)standard deviation for these columns.
mean_values=numeric_df.mean()
median_values=numeric_df.median()
mode_values=numeric_df.mode().iloc[0] #.iloc[0] gives the first mode value
std_dev=numeric_df.std()

print("\n Mean Values:\n", mean_values)
print("\n Median Values:", median_values)
print("\n Mode Values:", mode_values)
print("\n Standard Deviation:", std_dev)

#Provide a brief interpretation of these statistics:

print("\nBrief Interpretation:\n"
      "The mean gives the average value, showing the central tendency. "
      "\n The median shows the middle value, which is less affected by extreme values. "
      "\n The mode indicates the most frequently occurring value in each column. "
      "\n The standard deviation measures how spread out the data is around the mean. "
      "\n By the obtained output: (1) most SKUs have low sales volumes, but a few sell in high quantities; "
      "\n (2) Avg prices and sales values are highly skewed (mean > median > mode), indicating a few expensive products raise the average; "
      "\n (3) Discount rates are fairly consistent, but discount amounts vary widely due to high-value items; "
      "\n (4) Net sales shows high variability.")
#High variability = data points are widely spread from the mean → more inconsistency in the data.
#Low variability = data points are close to the mean → more consistency.
```

# Data Visualization:

#(1)Histogram: It helps to visualize the distribution of numeric columns.

```python
import matplotlib.pyplot as plt #matplotlib is a data visualization lib for the python programming,it used for creating wide variety of static, animated and interactive plots and charts for better understanding the data.
import pandas as pd
import numpy as np

#Load the dataset into a data analysis tool or programming environment.
file_path="/content/drive/MyDrive/Python excelr/sales_data_with_discounts.csv.xls"
df=pd.read_csv(file_path) #as ours is csv file , we are using read_csv()
numeric_df=df.select_dtypes(include=['int64','float64']) #here include=['int64','float64'] gives only columns with integer or floating point numbers

#plotting histogram
numeric_df.hist(figsize=(10,6),bins=15,edgecolor='black') # '.hist' is for the plotting the histogram, and the parameters are given accordingly
plt.suptitle('Histogram of Numeric Columns',fontsize=16) # '.suptitle' is main title for the histogram
plt.tight_layout() # '.tight_layout' we use this because x-axis labels, or y-axis labels can sometimes overlap and become unreadable.plt.tight_layout() solves this problem by a single call. It automatically calculates the optimal spacing between subplots, leaving enough room for all the labels and titles to be clearly visible.
plt.show()  # it displays the histogram.
```

Based on the histograms obtained above, most of the distributions for the sales data (Volume, Avg Price, Total Sales Value, Discount Amount, and Net Sales Value) are heavily right-skewed.
This indicates that the majority of transactions are small, but a few extremely large transactions act as outliers.
The presence of these outliers means that the mean is not a good representation of the typical value; the median would be a more accurate measure.
The Discount Rate distribution is different, it has two distinct discount strategies.

# (2)Box Plots: It shows median, quartiles, and outliers.

```python
import matplotlib.pyplot as plt #matplotlib is a data visualization lib for the python programming,it used for creating wide variety of static, animated and interactive plots and charts for better understanding the data.
import pandas as pd
import numpy as np

#Load the dataset into a data analysis tool or programming environment.
file_path="/content/drive/MyDrive/Python excelr/sales_data_with_discounts.csv.xls"
df=pd.read_csv(file_path) #as ours is csv file , we are using read_csv()
numeric_df=df.select_dtypes(include=['int64','float64']) #here include=['int64','float64'] gives only columns with integer or floating point numbers

#plotting boxplot
plt.figure(figsize=(10,6))
numeric_df.plot(kind='box')
plt.title("Box Plot of Numeric Columns:")
plt.xticks(rotation=45) # '.xticks' is for parameters, as they are big , rotation we help in viewing better
plt.tight_layout()
plt.show()
```

Based on the box plots that are obtained above, the sales data is heavily skewed and contains numerous outliers.
All columns except for Discount Rate show that the majority of values are concentrated at the low end, with many extreme high-value transactions. This confirms that the mean is not a reliable measure for these variables; the median is a better indicator of the typical value.

# (3)Bar Chart Analysis for Categorical Column: Identify categorical columns and plot bar charts.

```python
import matplotlib.pyplot as plt #matplotlib is a data visualization lib for the python programming,it used for creating wide variety of static, animated and interactive plots and charts for better understanding the data.
import pandas as pd
import numpy as np

#Load the dataset into a data analysis tool or programming environment.
file_path="/content/drive/MyDrive/Python excelr/sales_data_with_discounts.csv.xls"
df=pd.read_csv(file_path) #as ours is csv file , we are using read_csv()

# identifying categorical data
categorical_df = df.select_dtypes(include=['object'])  # 'object' data type in panda is used for text or strings, which are typically represent categorical data.

# Plot bar charts for each categorical column
for col in categorical_df.columns:
    plt.figure(figsize=(8,4))  # it creates a new figure everytime with those dimensions
    df[col].value_counts().plot(kind='bar', color='skyblue', edgecolor='black') # 'df[col].value_counts()'-counts how many times each unique category appears
    plt.title(f"Bar Chart of {col}")  # 'plot(kind='bar','-creates a bar chart with those counts
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

Based on Bar charts obtained :
1.Date,Day,SKU and Model:These categories have a uniform distribution ,meaning the counts for each unique value are nearly equal.
2.City:This has a single dominant category,as one city's sales count is significantly higher than the others.
3.BU:This shows a uniform distribution across mobiles , fmcg and lifestyle, indicating sales are balanced across these businesses.
4.Brand:The Brand distribution is non-uniform. 'jeera' is the most frequent brand and 'orange' and 'sumsang' are the least frequent brands.
5.This shows uniform distribution, with sales counts nearly identical for every product model. This means that no single model is more popular than another.
