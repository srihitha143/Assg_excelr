**1.Create a list of 5 integers. Perform and print the result of the following operations: append, extend, insert, remove, pop, clear, index, count, sort, and reverse.**

```python
nums = [] #an array for the list
try:
    print("Enter 5 numbers, separated by spaces:")  #user input and here 'item' is the variable
    user_input = input()
    nums = [int(item) for item in user_input.split()]
    print(f"Initial list: {nums}")
except ValueError:   #value error is like , if we give any other values or string or general error
    print("Invalid input.")
    exit() # exit if the initial list is invalid

# Append
try:
    x = int(input("Enter a number to append: "))
    nums.append(x)
    print(f"After append({x}): {nums}")
except ValueError:
    print("Invalid input.")

# Extend
try:
    e_str = input("Enter numbers to extend: ")
    nums.extend([int(item) for item in e_str.split()]) #here e_str.split() creats a list of strings , but again the int(item) converts them back into integer
    print(f"After extend: {nums}")
except ValueError:
    print("Invalid input.")

# Insert
try:
    pos = int(input("Enter position to insert: "))
    x = int(input("Enter a number to insert: "))
    nums.insert(pos, x)
    print(f"After insert({pos}, {x}): {nums}")
except (ValueError, IndexError):
    print("Invalid input.")

# Remove
try:
    x = int(input("Enter a number to remove: "))
    if x in nums:
        nums.remove(x)
        print(f"After remove({x}): {nums}")
    else:
        print(f"{x} is not in the list.")
except ValueError:
    print("Invalid input.")

# Pop an item by index
try:
    x = int(input(f"Enter an index to pop (0 to {len(nums)-1}): ")) #we know f is used for the considering the input in the braces also
    #here len(nums) gets the number of the items in the list, and the len(nums)-1 gets the last valid item
    popped_val = nums.pop(x)  #removes specific index and returns its value, that value is stored  in the variable we have given.
    print(f"After pop({x}): {nums}")
    print(f"Popped value: {popped_val}")
except (ValueError, IndexError):
    print("Invalid input.")

temp_nums = nums.copy() # because the previous list operations might have removed values, so we are creating a shallow copy of the list.

# index
try:
    x = int(input("Enter a number to find its index: "))
    print(f"Index of {x}: {temp_nums.index(x)}")
except ValueError:
    print(f"{x} is not in the list.")

# Count
try:
    x = int(input("Enter a number to count: "))
    print(f"Count of {x}: {temp_nums.count(x)}")
except ValueError:
    print("Invalid input.")

# Sort
temp_nums.sort()
print(f"After sort(): {temp_nums}")

# Reverse
temp_nums.reverse()
print(f"After reverse(): {temp_nums}")

# Clear the original list
nums.clear()
print(f"After clear(): {nums}")
```

**11.Create a small DataFrame manually with 10 rows and columns: 'Name'**
**'Age', 'City', and 'Salary'. Then:**
**- Use .info(), .describe()**
**- Select 'Name' and 'City' columns**
**- Drop 'City' column**
**- Fill any missing values in 'Salary' column with the mean**
**- Remove any duplicate rows**

```python
import numpy as np
import pandas as pd #we are importing pandas for this problem

#first we should create empty lists for the following columns
names=[]
ages=[]
cities=[]
salaries=[]

#now we should give the 10 inputs for these
print("enter the inputs:")

#as we should only give 10 inputs , we use:
for i in range(10):
  print(f"\nPerson {i+1}")  #i+1 because 10th input should be also included
  name=input("Name:")
  age=int(input("Age:"))
  city=input("City:")
  salary_input=input("Salary:")  #take salary as string first

  #As there should be missing salaries (as per the question), we should handle the missing salaries too
  if salary_input.strip()=="" or salary_input.lower()=="na":
      salary=np.nan #nan stands for Not a Number (used to represent missing values in NumPy/pandas. So if the input is empty or NA ,we store NaN.
  else:
    try:
      salary=float(salary_input)  # Convert to number only if valid
    except ValueError:
      print("invalid salary entered.")  # if user enters text instead of number
      salary=np.nan

  #now we should append these
  names.append(name)
  ages.append(age)
  cities.append(city)
  salaries.append(salary)

#now we should create a dataframe
df=pd.DataFrame({"Name":names,"Age":ages,"City":cities,"Salary":salaries})
print(df)

#now use.info() , this is used to get information
print("\n\n dataframe info")
print("\n",df.info())

#.describe() is used for description
print("\n\n dataframe description")
print("\n",df.describe())

#select means displaying the columns
print("\n\n name and city columns:")
print("\n",df[["Name", "City"]])

#drop city column , means delete a certain column
df_dropped=df.drop("City",axis=1)       #here we wrote axis=1, as axis tells pandas whether to drop columns or rows, axis=0 - drop by row; axis=1 - drop by column
print("\n\n after dropping city column:")
print("\n",df_dropped)

#fill any missing values in 'Salary' column with the mean
mean_salary=df["Salary"].mean()
df["Salary"].fillna(mean_salary, inplace=True)
print("\n\n after filling the missing values:")
print("\n",df)

#removing any duplicate rows
df_no_duplicates=df.drop_duplicates()
print("\n\n after removing duplicates:")
print("\n",df_no_duplicates)
```
